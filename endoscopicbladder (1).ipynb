```python
import tensorflow as tf
from tensorflow.keras import layers, models, callbacks
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.metrics import Precision, Recall, AUC
from tensorflow.keras.applications import VGG16
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, precision_recall_curve
import os
import matplotlib.pyplot as plt
import seaborn as sns
import json
from tensorflow.keras.preprocessing.image import load_img, img_to_array

# Chemin du dataset
base_path = "/kaggle/input/endoscopic-bladder/EndoscopicBladderTissue"

# Charger annotations.csv
annotations = pd.read_csv(os.path.join(base_path, "annotations.csv"))

# Mapper les classes en binaire
def map_label(tissue_type):
    if tissue_type in ['HGC', 'LGC']:
        return 1  # Massive
    elif tissue_type in ['NST', 'NTL']:
        return 0  # Non Massive
    else:
        raise ValueError(f"Tissue type inconnu : {tissue_type}")

annotations['binary_label'] = annotations['tissue type'].apply(map_label)



# Convertir binary_label en strings pour flow_from_dataframe
annotations['binary_label'] = annotations['binary_label'].astype(str)

# Vérifier les images manquantes
annotations['image_path'] = annotations.apply(
    lambda row: os.path.join(base_path, row['tissue type'], row['HLY']), axis=1
)
annotations = annotations[annotations['image_path'].apply(os.path.exists)]
print(f"Nombre d'images disponibles : {len(annotations)}")

# Paramètres
img_height, img_width = 224, 224
batch_size = 32  # Retour à 32
epochs = 20  # Augmenté



# Diviser les données (retour à 70/20/10)
train_data, temp_data = train_test_split(annotations, test_size=0.3, random_state=42)
val_data, test_data = train_test_split(temp_data, test_size=0.33, random_state=42)



# Augmentation de données
datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=10,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True,
    brightness_range=[0.8, 1.2],
    fill_mode='nearest'
)



# Générateurs
train_generator = datagen.flow_from_dataframe(
    train_data,
    x_col='image_path',
    y_col='binary_label',
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='binary'
)

val_generator = datagen.flow_from_dataframe(
    val_data,
    x_col='image_path',
    y_col='binary_label',
    target_size=(img_height, img_width),
    batch_size=16,
    class_mode='binary',
    shuffle=False
)

test_generator = datagen.flow_from_dataframe(
    test_data,
    x_col='image_path',
    y_col='binary_label',
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='binary',
    shuffle=False
)



# Construction du modèle (option 1 : modèle personnalisé)
# # Commenter cette section si tu veux utiliser VGG16
# model = models.Sequential([
#     layers.Input(shape=(img_height, img_width, 3)),
#     layers.Conv2D(32, (3, 3), activation='relu', padding='same'),
#     layers.MaxPooling2D((2, 2)),
#     layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
#     layers.MaxPooling2D((2, 2)),
#     layers.Conv2D(128, (3, 3), activation='relu', padding='same'),
#     layers.MaxPooling2D((2, 2)),
#     layers.Flatten(),
#     layers.Dense(128, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.01)),
#     layers.Dropout(0.5),
#     layers.Dense(1, activation='sigmoid')
# ])

# Option 2 : Transfer learning avec VGG16 (décommenter pour utiliser)
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(img_height, img_width, 3))
base_model.trainable = False
model = models.Sequential([
    base_model,
    layers.Flatten(),
    layers.Dense(128, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.01)),
    layers.Dropout(0.5),
    layers.Dense(1, activation='sigmoid')
])

# Construction du modèle avec EfficientNetB0
# base_model = EfficientNetB0(weights='imagenet', include_top=False, input_shape=(img_height, img_width, 3))
# base_model.trainable = False  # Freeze the base model

# model = models.Sequential([
#     base_model,
#     layers.Flatten(),
#     layers.Dense(128, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.01)),
#     layers.Dropout(0.5),
#     layers.Dense(1, activation='sigmoid')
# ])


# Construction du modèle avec ResNet50
# base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(img_height, img_width, 3))
# base_model.trainable = False  # Freeze the base model

# model = models.Sequential([
#     base_model,
#     layers.Flatten(),
#     layers.Dense(128, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.01)),
#     layers.Dropout(0.5),
#     layers.Dense(1, activation='sigmoid')
# ])



# Compilation
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
    loss='binary_crossentropy',
    metrics=['accuracy', Precision(name='precision'), Recall(name='recall'), AUC(name='auc')]
)
model.summary()



# Early Stopping
early_stopping = callbacks.EarlyStopping(monitor='val_loss', patience=7, restore_best_weights=True)

# Calculer les poids des classes (rééquilibrage)
class_weights = {
    0: (len(train_data) / (2 * np.sum(train_data['binary_label'] == '0'))),
    1: (len(train_data) / (2 * np.sum(train_data['binary_label'] == '1'))) * 1.2  # Réduit à 1.2
}
print("Poids des classes ajustés :", class_weights)



# Entraînement
history = model.fit(
    train_generator,
    epochs=epochs,
    validation_data=val_generator,
    callbacks=[early_stopping],
    class_weight=class_weights
)

# Sauvegarder l'historique
with open('training_history.json', 'w') as f:
    json.dump(history.history, f)
print("Historique d'entraînement sauvegardé sous 'training_history.json'")

# Évaluation sur le set de test
metrics = model.evaluate(test_generator)
test_loss = metrics[0]
test_accuracy = metrics[1]
test_precision = metrics[2]
test_recall = metrics[3]
test_auc = metrics[4]
print(f"Test Loss: {test_loss:.4f}")
print(f"Test Accuracy: {test_accuracy:.4f}")
print(f"Test Precision: {test_precision:.4f}")
print(f"Test Recall: {test_recall:.4f}")
print(f"Test AUC: {test_auc:.4f}")

# Courbes d'entraînement
plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.subplot(2, 2, 3)
plt.plot(history.history['recall'], label='Training Recall')
plt.plot(history.history['val_recall'], label='Validation Recall')
plt.title('Recall')
plt.xlabel('Epoch')
plt.ylabel('Recall')
plt.legend()

plt.subplot(2, 2, 4)
plt.plot(history.history['auc'], label='Training AUC')
plt.plot(history.history['val_auc'], label='Validation AUC')
plt.title('AUC')
plt.xlabel('Epoch')
plt.ylabel('AUC')
plt.legend()

plt.tight_layout()
plt.show()



# Prédictions et rapport de classification (Validation)
y_pred = model.predict(val_generator)
y_pred_binary = (y_pred > 0.5).astype(int)
y_val = val_data['binary_label'].values.astype(int)

# Créer un DataFrame pour les prédictions de validation
val_predictions_df = pd.DataFrame({
    'Image Path': val_data['image_path'].values,
    'True Label': ['Massive' if label == 1 else 'Non Massive' for label in y_val],
    'Predicted Label': ['Massive' if pred == 1 else 'Non Massive' for pred in y_pred_binary.flatten()],
    'Prediction Probability': y_pred.flatten()
})
print("\nPrédictions sur le set de validation :")
print(val_predictions_df)

# Sauvegarder les prédictions de validation dans un fichier CSV
val_predictions_df.to_csv('validation_predictions.csv', index=False)
print("Prédictions de validation sauvegardées sous 'validation_predictions.csv'")

# Rapport de classification (Validation)
print("\nRapport de classification sur le set de validation :")
print(classification_report(y_val, y_pred_binary, target_names=["Non Massive", "Massive"]))

# Prédictions et rapport de classification (Test)
y_test_pred = model.predict(test_generator)
y_test_pred_binary = (y_test_pred > 0.5).astype(int)
y_test = test_data['binary_label'].values.astype(int)

# Créer un DataFrame pour les prédictions de test
test_predictions_df = pd.DataFrame({
    'Image Path': test_data['image_path'].values,
    'True Label': ['Massive' if label == 1 else 'Non Massive' for label in y_test],
    'Predicted Label': ['Massive' if pred == 1 else 'Non Massive' for pred in y_test_pred_binary.flatten()],
    'Prediction Probability': y_test_pred.flatten()
})
print("\nPrédictions sur le set de test :")
print(test_predictions_df)

# Sauvegarder les prédictions de test dans un fichier CSV
test_predictions_df.to_csv('test_predictions.csv', index=False)
print("Prédictions de test sauvegardées sous 'test_predictions.csv'")

# Rapport de classification (Test, seuil par défaut)
print("\nRapport de classification sur le set de test :")
print(classification_report(y_test, y_test_pred_binary, target_names=["Non Massive", "Massive"]))

# Matrice de confusion (seuil par défaut)
cm = confusion_matrix(y_test, y_test_pred_binary)
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Non Massive', 'Massive'], yticklabels=['Non Massive', 'Massive'])
plt.title('Matrice de confusion (Test, seuil = 0.5)')
plt.xlabel('Prédit')
plt.ylabel('Réel')
plt.show()

# Visualisation de la distribution des probabilités
plt.figure(figsize=(10, 6))
plt.hist(y_pred[y_val == 1], bins=50, alpha=0.5, label='Massive (True)', color='red')
plt.hist(y_pred[y_val == 0], bins=50, alpha=0.5, label='Non Massive (True)', color='blue')
plt.axvline(x=0.5, color='black', linestyle='--', label='Seuil par défaut (0.5)')
plt.title('Distribution des probabilités prédites (Validation)')
plt.xlabel('Probabilité prédite')
plt.ylabel('Nombre d\'images')
plt.legend()
plt.show()

# Ajustement du seuil (maximisation du F1-score et recall >= 97%)
precisions, recalls, thresholds = precision_recall_curve(y_val, y_pred)

# Seuil pour recall >= 97%
optimal_threshold_recall = thresholds[np.argmax(recalls >= 0.97)]
print(f"Seuil optimal pour un recall >= 97% : {optimal_threshold_recall:.2f}")
y_pred_binary_opt_recall = (y_pred > optimal_threshold_recall).astype(int)
print("Rapport de classification avec seuil ajusté (recall >= 97%, validation) :")
print(classification_report(y_val, y_pred_binary_opt_recall, target_names=["Non Massive", "Massive"]))

# Seuil pour maximiser le F1-score
f1_scores = 2 * (precisions * recalls) / (precisions + recalls + 1e-10)
optimal_idx_f1 = np.argmax(f1_scores)
optimal_threshold_f1 = thresholds[optimal_idx_f1]
print(f"\nSeuil optimal pour maximiser le F1-score : {optimal_threshold_f1:.2f}")
y_pred_binary_opt_f1 = (y_pred > optimal_threshold_f1).astype(int)
print("Rapport de classification avec seuil ajusté (max F1-score, validation) :")
print(classification_report(y_val, y_pred_binary_opt_f1, target_names=["Non Massive", "Massive"]))

# Appliquer le seuil optimal (F1-score) sur le set de test
y_test_pred_binary_opt = (y_test_pred > optimal_threshold_f1).astype(int)
print("\nRapport de classification sur le set de test avec seuil ajusté (F1-score) :")
print(classification_report(y_test, y_test_pred_binary_opt, target_names=["Non Massive", "Massive"]))

# Nouvelle matrice de confusion avec seuil ajusté
cm_opt = confusion_matrix(y_test, y_test_pred_binary_opt)
plt.figure(figsize=(6, 4))
sns.heatmap(cm_opt, annot=True, fmt='d', cmap='Blues', xticklabels=['Non Massive', 'Massive'], yticklabels=['Non Massive', 'Massive'])
plt.title(f'Matrice de confusion (Test, seuil = {optimal_threshold_f1:.2f})')
plt.xlabel('Prédit')
plt.ylabel('Réel')
plt.show()

# Visualisation des prédictions
plt.figure(figsize=(15, 6))
for i, idx in enumerate(range(min(5, len(val_data)))):
    plt.subplot(1, 5, i+1)
    img = plt.imread(val_data['image_path'].iloc[idx])
    plt.imshow(img)
    true_label = "Massive" if y_val[idx] == 1 else "Non Massive"
    pred_label = "Massive" if y_pred_binary[idx] == 1 else "Non Massive"
    plt.title(f"True: {true_label}\nPred: {pred_label}\nProb: {y_pred[idx][0]:.2f}")
    plt.axis('off')
plt.tight_layout()
plt.show()

# Sauvegarde du modèle
model.save('cnn_bladder_massive.keras')
print("Modèle sauvegardé sous 'cnn_bladder_massive.keras'")
```
